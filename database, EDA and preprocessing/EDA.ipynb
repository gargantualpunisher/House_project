```python
import pandas as pd 
import seaborn as sns 
import matplotlib.pyplot as plt
%matplotlib inline
```

# Property type(Houses and Flats)

```python
file['property_type'].value_counts()
```

```python
file['property_type'].value_counts().plot(kind='bar')
```

# Price of property on basis of type vs price_per_sqft

```python
top_10 = file[file['price_per_sqft'] > 23712]
```

```python
society_price_mean = top_10.groupby('society')['price_per_sqft'].mean().sort_values(ascending=False)

plt.figure(figsize=(8, 5))
society_price_mean[:10].plot(kind='barh')
plt.title('Top 10 Societies by Average Price per Sqft')
plt.xlabel('Society')
plt.ylabel('Average Price per Sqft')
plt.show()


```

# top_10 propety Locationwise

```python
society_price_mean = top_10.groupby('Location')['price_per_sqft'].mean().sort_values(ascending=False)

plt.figure(figsize=(8, 5))
society_price_mean[:10].plot(kind='barh')
plt.title('Top 10 Location by Average Price per Sqft')
plt.xlabel('Society')
plt.ylabel('Average Price per Sqft')
plt.show()
```

# price vs Location

```python
society_price_mean = top_10.groupby('Location')['price1'].mean().sort_values(ascending=False)

plt.figure(figsize=(8, 5))
society_price_mean[:10].plot(kind='barh')
plt.title('Top 10 Location by Average Price per Sqft')
plt.xlabel('Society')
plt.ylabel('Average Price per Sqft')
plt.show()
```

# price vs society

```python
society_price_mean = top_10.groupby('society')['price1'].mean().sort_values(ascending=False)

plt.figure(figsize=(8, 5))
society_price_mean[:10].plot(kind='barh')
plt.title('Top 10 Location by Average Price per Sqft')
plt.xlabel('Society')
plt.ylabel('Average Price per Sqft')
plt.show()
```

# price vs Property type

# bedRoom vs property_type

```python
sns.histplot(data=house, x='bedRoom', hue='property_type', color='green', element='step')

sns.histplot(data=flat, x='bedRoom', hue='property_type', color='red', element='step')

plt.xlabel('Number of Bedrooms')
plt.ylabel('Count')
plt.title('Distribution of Bedrooms by Property Type')
plt.legend(title='Property Type', labels=['House', 'Flat'])

plt.show()
```
