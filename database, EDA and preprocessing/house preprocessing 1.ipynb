```python
pattern = r'Plot area \d+\d.\d.\d+.\d+ sq.m.'
plot = []
```

```python
import re

# Assuming app1['areaWithType'] is a list of values
for i in app1['areaWithType']:
    if isinstance(i, str) and 'Plot' in i:
        match = re.search(pattern, i)
        if match:
            plot.append(match.group())
        
            
        else:
            plot.append(0)
            
    
    else:
        plot.append(0)
```

```python
built_up_area_pattern = r'Built Up area: (\d+(\.\d+)?) sq\.yards \((\d+(\.\d+)?) sq\.m\.\)'
build = []
```

```python
for i in app1['areaWithType']:
    if isinstance(i, str) and 'Built Up area' in i:
        match = re.search(built_up_area_pattern, i)
        if match:
            build.append(match.group())
        
            
        else:
            build.append(0)
            
    
    else:
        build.append(0)
```

```python
build
```

```python
# Carpet area: 460 sq.yards (384.62 sq.m.)
```

```python
carpet_pattern = r'Carpet area: (\d+(\.\d+)?) sq\.yards \((\d+(\.\d+)?) sq\.m\.\)'
carpet = []
```

```python
for i in app1['areaWithType']:
    if isinstance(i, str) and 'Built Up area' in i:
        match = re.search(carpet_pattern, i)
        if match:
            carpet.append(match.group())
        
            
        else:
            carpet.append(0)
            
    
    else:
        carpet.append(0)
```

```python
new['carpet'] = carpet
new['build'] = build
new['plot'] = plot
```

```python
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
from sklearn.preprocessing import StandardScaler

sa = StandardScaler()
sacaled = pd.DataFrame(sa.fit_transform(fa))
```

```python
k_values = range(1, 11)
inertia_values = []

# Calculate the within-cluster sum of squares (inertia) for each k
for k in k_values:
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(sacaled)
    inertia_values.append(kmeans.inertia_)

# Plot the elbow curve
plt.figure(figsize=(8, 4))
plt.plot(k_values, inertia_values, marker='o', linestyle='-')
plt.title('Elbow Method for Optimal K')
plt.xlabel('Number of Clusters (K)')
plt.ylabel('Within-Cluster Sum of Squares (Inertia)')
plt.grid(True)
plt.show()
```

```python
kmeans = KMeans(n_clusters=3)

# Fit KMeans to the data
kmeans.fit(sacaled)

# Get cluster labels and cluster centers
labels = kmeans.labels_
centers = kmeans.cluster_centers_
clusters = kmeans.predict(fa)
```

```python
mlb = MultiLabelBinarizer()
```

```python
matrix = mlb.fit_transform(app['features_list'])
```

```python
matrix
```

```python
feature_matrix = pd.DataFrame(matrix, columns=mlb.classes_)
```

```python
wcss = []

for i in range(1,11):
    kmeans = KMeans(n_clusters=i, random_state=42)
    kmeans.fit(feature_matrix)
    wcss.append(kmeans.inertia_)
```

```python
k = range(1,11)
```

```python
plt.figure(figsize=(8, 4))
plt.plot(k, wcss, marker='o', linestyle='-')
plt.title('Elbow Method for Optimal K')
plt.xlabel('Number of Clusters (K)')
plt.ylabel('Within-Cluster Sum of Squares (Inertia)')
plt.grid(True)
plt.show()
```
